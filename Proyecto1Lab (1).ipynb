## Marco Teórico

### 1. Regresión lineal
La regresión lineal es un método estadístico que busca encontrar una relación entre una variable dependiente (Y) y una o más variables independientes (X).  
La idea es ajustar una línea recta que mejor explique cómo cambian los valores de Y según los cambios en X.  
Ejemplo: predecir el nivel de felicidad de un país en función de su PIB per cápita.

### 2. Regresión polinomial
Cuando la relación entre las variables no es lineal, se puede usar un polinomio (por ejemplo, cuadrático o cúbico).  
La regresión polinomial agrega potencias de X (X², X³, etc.) para capturar curvas.  
Ejemplo: la relación entre expectativa de vida y felicidad puede ser curva (crece rápido al inicio y luego se estabiliza).

### 3. Interacción de factores
La interacción ocurre cuando el efecto de una variable depende del valor de otra.  
Ejemplo: el impacto del PIB en la felicidad puede ser diferente según el nivel de apoyo social; es decir, PIB y apoyo social interactúan.

### 4. Significancia de factores
No todas las variables explican de forma importante a la variable dependiente.  
La significancia estadística (p-value) nos indica si un factor realmente aporta al modelo o si su efecto es tan pequeño que podría ser casualidad.  
Ejemplo: verificar si la percepción de corrupción tiene un efecto significativo en la felicidad.

### 5. Regularización (Ridge, Lasso, ElasticNet)
La regularización ayuda a mejorar los modelos cuando hay muchas variables o cuando los datos generan sobreajuste.  
- **Ridge**: penaliza los coeficientes grandes pero nunca los lleva a cero.  
- **Lasso**: además de penalizar, puede llevar algunos coeficientes exactamente a cero (sirve para seleccionar variables).  
- **ElasticNet**: combina Ridge y Lasso.  
Ejemplo: al usar muchas variables (PIB, apoyo social, generosidad, corrupción, etc.), la regularización evita que el modelo se complique demasiado y pierda capacidad de generalización.

## Análisis del dataset

 ### ¿De dónde viene?

El dataset proviene de Kaggle y corresponde a los World Happiness Reports, un conjunto de datos que recopila indicadores de bienestar subjetivo y condiciones socioeconómicas de distintos países.

 ### ¿Qué contiene?

- Life Ladder: índice de felicidad.
- Log GDP per capita: PIB per cápita en logaritmo.
- Social support: percepción de apoyo social.
- Healthy life expectancy at birth: expectativa de vida saludable.
- Freedom to make life choices: libertad para decidir sobre la vida.
- Generosity: disposición a la generosidad.
- Perceptions of corruption: percepción de corrupción.
- Positive affect y Negative affect: emociones reportadas por la población.

 ### ¿Qué información dan las muestras?
Las muestras reflejan los valores de 165 países y su evolución del bienestar subjetivo a medida que transcurren los años. Con ello, hacemos referencia a una felicidad contruída con factores socieconómicos, políticos e institucionales. 


### ¿Qué se quiere analizar?

Buscamos llegar a la conclusión de que no siempre la felicidad depende de uno mismo, sino que hay muchos factores alrededor nuestro que pueden llegar a tener mucha relevancia en la cantidad de presencia de esta emoción en nuestras vidas. Queremos observar cuáles variables son más significativas, así como el observar cuáles son los países que tienen un nivel más bajo de este parámetro social. 


### ¿Qué variables se tienen que transformar para poder usarse en un modelo de regresión?

Estas variables incluyen principalmente la que alberga los países (Country Name) ya que, al ser una variable categórica, debemos convertirla en dummies para lograr su análisis y para que pueda ser incluída dentro del modelo. Asimismo, este proceso genera más de 160 columnas (uno por cada país), lo que complica el análisis, puesto que se tienen coeficientes y resulta difícil identificar cuáles son estadísticamente significativos con el modelo de OLS. Por ello, en algunos casos es necesario excluir temporalmente la variable país al buscar significancia de los predictores principales y luego reincorporarla al modelo, dado que consideramos que el país influye de gran manera en los niveles de felicidad.
Además, fue necesario escalar las variables para que puedan ser comparadas igualitariamente. Esto a razón de que las variables están dadas en diferentes unidades, algunas en decimales y otras en decenas. De esta manera, al hacer escalamiento, colocamos todos los datos en una misma escala, vaya la redundancia. 

### ¿Qué resultado se podría encontrar al realizar una regresión?

- El modelo permitirá explicar el cómo la felicidad de un país está relacionada con factores económicos, sociales, políticos y emocionales. Además, nos permitirá evaluar el peso y la importancia de cada factor mediante sus coeficientes y p-values.
- Podremos comprobar que el PIB per cápita y la expectativa de vida son los predictores más fuertes. Así como corroborar que la corrupción pueda estar asociada de manera negativa con la felicidad.

# Modelo propuesto

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from scipy import stats
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("World-happiness-report-updated_2024.csv", encoding='latin1')
data.head()
```

## Modelo 1

![Gráfico de resultados](1.jpeg)

```python
df = data.dropna().drop_duplicates()

x1 = df.drop(["Life Ladder", "Country name"], axis=1)
y = df[["Life Ladder"]]
```

#### Análisis de significancia de factores para modelo sin penalización.

#### OLS

```python
X1 = np.column_stack([x1])

X_1 = sm.add_constant(X1)
ols = sm.OLS(y, X_1).fit()
print(ols.summary())
```

#### Antes de comenzar con los siguientes modelos, vamos a eliminar la variable de "negative affect" puesto que tiene un p_value mayor a 0.05, lo cual nos dice que no es significativo para el modelo.

```python
df = pd.get_dummies(df, columns=["Country name"], drop_first=True)

x = df.drop(["Life Ladder", "Negative affect"], axis=1)
```

**Linear Regression**

```python
scaler = StandardScaler().fit(x)
x_scaled = scaler.transform(x)
```

```python
lin = LinearRegression()
lin.fit(x_scaled, y)
y_pred_lin = lin.predict(x_scaled)
r2_lin = r2_score(y, y_pred_lin)
r2_lin
```

#### **Ridge**

```python
rd = Ridge(alpha=1)
rd.fit(x_scaled, y)
rd.intercept_
```

```python
y_pred_rd = rd.predict(x_scaled)
r2_rd = r2_score(y, y_pred_rd)
r2_rd
```

#### **Lasso**

```python
ls = Lasso(alpha=0.1)
ls.fit(x_scaled, y)
ls.intercept_
```

```python
y_pred_ls = ls.predict(x_scaled)
r2_ls = r2_score(y, y_pred_ls)
r2_ls
```

#### **ElasticNet**

```python
en = ElasticNet(alpha=0.1, l1_ratio=0.5)
en.fit(x_scaled, y)
ls.intercept_
```

```python
y_pred_en = en.predict(x_scaled)
r2_en = r2_score(y, y_pred_en)
r2_en
```

#### **Visualizaciones Modelo 1**

```python
plt.figure(figsize=(12,6))

order = np.argsort(y.values.ravel())
y_sorted = y.values.ravel()[order]


plt.plot(y_pred_lin[order], label=f"Linear (R²={r2_lin:.3f})", linestyle="--")
plt.plot(y_pred_rd[order], label=f"Ridge (R²={r2_rd:.3f})", linestyle="--")
plt.plot(y_pred_ls[order], label=f"Lasso (R²={r2_ls:.3f})", linestyle="--")
plt.plot(y_pred_en[order], label=f"ElasticNet (R²={r2_en:.3f})", linestyle="--")

plt.plot(y_sorted, label="Datos reales", linewidth=2, color="black")

plt.title("Comparación de modelos lineales sobre Life Ladder")
plt.xlabel("Observaciones (ordenadas por valor real)")
plt.ylabel("Life Ladder")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Valores de R² para cada modelo
r2 = [r2_lin, r2_rd, r2_ls, r2_en]

modelos = ["Linear", "Ridge", "Lasso", "ElasticNet"]
x = np.arange(len(modelos))  # posiciones en X
width = 0.35  # ancho de cada barra

plt.figure(figsize=(10,6))
# Barras para train y test
plt.bar(x - width/2, r2, width, label="R2", color="skyblue", edgecolor="black")

# Ejes y títulos
plt.ylabel("R²")
plt.xlabel("Modelos")
plt.title("Comparación de R² en Train y Test modelo 1")
plt.xticks(x, modelos)  # poner nombres de los modelos en el eje X
plt.ylim(0, 1)          # opcional, si quieres limitar entre 0 y 1
plt.legend()
plt.show()
```

## Conclusión

En este proyecto buscamos predecir el nivel de felicidad de un país (Life Ladder) utilizando distintos modelos de regresión. 

Desde el inicio enfrentamos problemas importantes con la variable Country Name, ya que contenía 165 países distintos. Al transformarla en variables dummy el número de coeficientes se volvió inmanejable y no era viable eliminar países en función de su significancia estadística. Por este motivo, en el primer modelo con OLS decidimos excluir esta variable y concentrarnos en los demás factores.

En el primer modelo lineal, observamos que casi todas las variables fueron significativas, salvo Negative affect, que se eliminó del modelo. Los resultados mostraron que Log GDP per capita fue el predictor con mayor peso, mientras que otras variables como Social support y Perceptions of Corruption también tuvieron relevancia. Linear Regression y Ridge con todos los datos alcanzaron un 𝑅2 cercano a 0.91, confirmando su buen desempeño. Con cross-validation, al dividir los datos en 60% entrenamiento y 40% prueba, los resultados se estabilizaron con un 𝑅2 de alrededor de 0.876, mostrando un ajuste sólido sin exceso de sobreajuste.

En nuestro segundo modelo, de tipo cuadrático, se buscaba capturar relaciones no lineales entre las variables. Si bien algunos términos al cuadrado fueron significativos, el modelo mostró un problema claro de sobreajuste: al entrenar con todos los datos, Linear y Ridge alcanzaron un 𝑅2=1. Con cross-validation los resultados bajaron a un nivel razonable, lo que confirmó que el modelo original estaba memorizando el dataset en lugar de generalizarlo.

En el tercer modelo incorporamos una interacción entre Log GDP per capita y Social support, bajo la hipótesis de que el efecto combinado de estos factores podía ser relevante. El coeficiente de la interacción resultó altamente significativo y el modelo mejoró en términos de ajuste. Sin embargo, nuevamente apareció el problema de sobreajuste: con todos los datos, Linear y Ridge entregaron 𝑅2=1. Incluso con cross-validation los resultados se mantuvieron sospechosamente altos (𝑅2≈0.99) lo que indica que el modelo podría estar capturando patrones demasiado específicos del dataset.

Una vez sabiendo esto los modelos de Linear Regression y Ridge fueron los más consistentes, con alto poder explicativo y menor riesgo de sobreajuste. Los modelos penalizados como Lasso y ElasticNet ofrecieron menor precisión, pero aportaron simplificación al reducir o encoger coeficientes. El uso de modelos cuadráticos e interacciones mostró potencial, pero también el riesgo de generar sobreajuste. 
En términos generales, los factores más relevantes para explicar la felicidad de un país fueron el PIB per cápita, el apoyo social, la salud y ciertas variables de percepción como la generosidad y la corrupción, así como la interacción entre PIB y apoyo social. 

Ahora nos damos cuenta que no siempre la felicidad depende de nosotros mismos. Hay muchos factores en nuestra sociedad, comunidad o país que son importantes a considerar en cuanto al cómo nos sentimos. Tenemos la percepción de que muchas veces pensamos  que estar felices es cuestión de uno mismo, de ser optimistas, de ver la vida desde un enfoque diferente, de ser fuertes, pero no siempre es así. Hay ocasiones que por el simple hecho de pertenecer a un país ya se crea una idea automática de que tanta felicidad existirá en la persona. Razonoes geográficas, climáticas, sociales, influyen en toda esta contrucción de nuestra propia felicidad. Por ello la importancia de que un país tenga en mente estos puntos, por el papel tan clave con el que juegan. Enfocándonos estos factores como el PIB per cápita y el apoyo social, se es posible mejorar la calidad de vida de las personas. Y al final del día es como un ciclo, entre mejor se sienta una persona, mejor rendimiento tendrá, más trabajará, mayor aportará a la sociedad y por lo tanto al país en general.
